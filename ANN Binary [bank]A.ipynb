```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["job"] = le.fit_transform(data["job"])
data["housing"] = le.fit_transform(data["housing"])
data["loan"] = le.fit_transform(data["loan"])
data["contact"] = le.fit_transform(data["contact"])
data["month"] = le.fit_transform(data["month"])
data["poutcome"] = le.fit_transform(data["poutcome"])
data["deposit"] = le.fit_transform(data["deposit"])
```

```python
x = data.iloc[:,0:13].values
y = data.iloc[:,13].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
b = Sequential()
```

```python
b.add(Dense(13,kernel_initializer="random_uniform",activation="relu"))
```

```python
b.add(Dense(26,kernel_initializer="random_uniform",activation="relu"))
```

```python
b.add(Dense(1,kernel_initializer="random_uniform",activation="sigmoid"))
```

```python
b.compile("adam","binary_crossentropy",metrics =["accuracy"])
```

```python
b.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
ypred = b.predict(x_test)
```
