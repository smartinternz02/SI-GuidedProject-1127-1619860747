```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([("oh",OneHotEncoder(),[0,1,3,4,10])],remainder ="passthrough")
x=ct.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train = np.asarray(x_train).astype('float32')
x_test = np.asarray(x_test).astype('float32')
y_train = np.asarray(y_train).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ls = Sequential()
```

```python
ls.add(Dense(11,kernel_initializer="random_uniform",activation="relu"))
```

```python
ls.add(Dense(22,kernel_initializer="random_uniform",activation="relu"))
```

```python
ls.add(Dense(2,kernel_initializer="random_uniform",activation="softmax"))
```

```python
ls.compile("adam","categorical_crossentropy",metrics =["accuracy"])
```

```python
ls.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
ypred = ls.predict(x_test)
```
