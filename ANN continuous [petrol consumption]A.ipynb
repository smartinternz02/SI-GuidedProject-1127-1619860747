```python
x = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1000)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
pc = Sequential()
```

```python
pc.add(Dense(4,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
pc.add(Dense(8,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
pc.add(Dense(1,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
pc.compile("adam","mean_squared_error",metrics = ["MSE"])
```

```python
pc.fit(x_train,y_train,batch_size = 32,epochs = 10)
```

```python
ypred = pc.predict(x_test)
ypred
```
